```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import os
import pandas as pd
import math
import re
import contractions

ROOT = r"C:\\Users\\USER\\Downloads\\Devraj"
DATA_DIR = rf"{ROOT}\paraphrased_data"
MBTI_RAW_CSV_PATH = os.path.join(DATA_DIR, "cleaned_post_2302.csv")
PARA_DIR = rf"{DATA_DIR}\paraphrased_5"
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

def paraphrase_text(input_text, model_name="gpt2", max_length=50):
    # Load pre-trained GPT-2 model and tokenizer
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # Tokenize input text
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # Generate attention mask
    attention_mask = torch.ones_like(input_ids)

    # Generate paraphrased text with appropriate parameters
    output = model.generate(
        input_ids,
        max_length=max_length,
        num_beams=5,
        no_repeat_ngram_size=2,
        temperature=0.7,
        top_k=50,
        top_p=0.95,
        do_sample=True,
        attention_mask=attention_mask,
        pad_token_id=tokenizer.pad_token_id
    )

    # Decode the generated output
    paraphrased_text = tokenizer.decode(output[0], skip_special_tokens=True)

    return paraphrased_text

# Example usage
df = pd.read_csv(MBTI_RAW_CSV_PATH)

# Iterate over each row and create/append paraphrased CSV files
for index, row in df.iterrows():
    type_name = row['type']
    cleaned_post = row['cleaned_post']

    # Create a new column for paraphrased text and save it immediately for each row
    df.at[index, "Paraphrased Text"] = ""
    
    # Split the text and paraphrase it
    original_text_list = cleaned_post.split("|||")
    print("No of Sentences in the Posts:",len(original_text_list))
    number_of_sent = 5
    max_length = 200
    paraphrased_text_list = []
    end = math.floor(len(original_text_list)/number_of_sent)
    for i in range(0,end,number_of_sent):
        text = ""
        if len(original_text_list[i]) <= 3:
            continue
        for j in range(number_of_sent):
            text = text + original_text_list[i+j]
        
        try:
            paraphrased_text = paraphrase_text(text, max_length=max_length)

        except RuntimeError as e:
            #Calculate max_length based on the size of the existing tensor
            existing_tensor_size = len(tokenizer.encode(text, return_tensors="pt")[0])
            max_length = existing_tensor_size + 10  # Add a buffer (adjust as needed)
            paraphrased_text = paraphrase_text(text, max_length=max_length)

        paraphrased_text = paraphrased_text.replace('Reply ','')
        paraphrased_text = paraphrased_text.replace('Report Post','')
        paraphrased_text = paraphrased_text.replace('Anonymous 05/27/15 (Tue) 07:', '')
        paraphrased_text = re.sub(r'Anonymous \d{2}/\d{2}/\d{2} \(\w+\) \d{2}:\d{2}:\d{2} [APMapm]{2} No\. \d+ >>\d+', '', paraphrased_text)
        paraphrased_text = contractions.fix(paraphrased_text)
        paraphrased_text = re.sub(r' +', ' ', paraphrased_text)
        paraphrased_text_list.append(paraphrased_text)

    # Save the paraphrased text along with type to a CSV file
    paraphrased_df = pd.DataFrame({'Type': [type_name]*len(paraphrased_text_list),
                                   'Paraphrased Text': paraphrased_text_list})
    output_csv_path = os.path.join(PARA_DIR, f"{number_of_sent}_sentence_{type_name}.csv")

    if os.path.exists(output_csv_path):
        paraphrased_df.to_csv(output_csv_path, mode='a', header=False, index=False)
    else:
        paraphrased_df.to_csv(output_csv_path, index=False)

    print(f"Processed and saved paraphrased data for type: {type_name}")
```
